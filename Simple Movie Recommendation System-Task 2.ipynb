```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies = pd.read_csv("/Users/priyangshubagchi/Downloads/archive (7)/tmdb_5000_movies.csv")
credits = pd.read_csv("/Users/priyangshubagchi/Downloads/archive (7)/tmdb_5000_credits.csv")
print(movies.shape)
movies.head() 
```

```python
movies = movies[['title', 'overview', 'genres']]

movies.head() 
```

```python
movies.dropna(inplace=True)

movies.isnull().sum()
```

```python
movies['tags'] = movies['overview']

movies[['title', 'tags']].head()
```

```python
tfidf = TfidfVectorizer(
    stop_words='english',
    max_features=5000
)

vectors = tfidf.fit_transform(movies['tags']) 
```

```python
similarity = cosine_similarity(vectors)

print(similarity.shape) 
```

```python
def recommend(movie_name):

    movie_name = movie_name.lower()

    matches = movies[
        movies['title'].str.lower() == movie_name
    ]

    if matches.empty:
        print("Movie not found!")
        return

    index = matches.index[0]

    distances = list(
        enumerate(similarity[index])
    )

    distances = sorted(
        distances,
        key=lambda x: x[1],
        reverse=True
    )

    print("\nTop 5 Recommendations:\n")

    for movie in distances[1:6]:
        print(
            movies.iloc[movie[0]].title
        ) 
```

```python
recommend("Avatar") 
```

```python
recommend("The Dark Knight")
```

```python
movie = input("Enter Movie Name: ")

recommend(movie) 
```
